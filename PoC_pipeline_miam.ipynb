```python
import pymupdf

# init value
full_path_input_user = ""
relative_path_input_user = "./data/input_user/"

pdf_doc = "./data/input_user/file_01.pdf"
```

```python
pdf_doc = pymupdf.open(pdf_doc)
print(pdf_doc)

out = open("output.txt", "wb")

for page in pdf_doc:
    text = page.get_text().encode("utf-8")
    out.write(text)
    out.write(bytes((12,)))
out.close()
```
